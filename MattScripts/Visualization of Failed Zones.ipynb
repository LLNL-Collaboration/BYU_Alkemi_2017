# Failed Zone Visualization

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
import random
```

```python
def my_read_file(num):
    '''This function is for reading in the csv's and panda data frames
    '''
    filename = "fail_data_" + str(num) + ".csv"
    data = pd.read_csv(filename)
    return data
```

```python
#colors for the graphs
colors = ['red','blue','green','orange','purple','yellow',\
         'turquoise','gray','gold','aquamarine','pink','maroon',\
         'brown','indigo']
for i in range(0,16): #loops through all 16 features of the data

    color_choice = random.randint(0,len(colors)-1)
    color_choice
    for run in range(0,11): #loops through the intial 10 failure runs 
        df = my_read_file(run)

    
        x = range(0,df.shape[0])
        col = df.columns[i]
        y = df[col]



        plt.plot(x,y,'-o',color=colors[color_choice])
        plt.xlabel("Time")
        plt.ylabel(col)
        plotname=(df.columns[i] + "for Run:  " + str(run))
        plt.title(plotname)
        plt.show()
```
