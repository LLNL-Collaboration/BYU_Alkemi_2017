# Analysis of Failed Zones

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

Here you can input the desired run and it will give us the data for all cycles of the failed zone in that run.

```python
def my_read_file(num):
    filename = "fail_data_" + str(num) + ".csv"
    data = pd.read_csv(filename)
    return data
    
num = input("Input Run: ",)
my_read_file(num).head()
```

```python
#make a plot for each of the spikes
for i in range(0,11):
    file = "fail_data_" + str(i) + ".csv"
    data = pd.read_csv(file)
    
    x = range(0,data.shape[0])
    y = data["oddy"]
    
    
    colors = ['red','blue','green','orange','purple','yellow',\
             'turquoise','gray','gold','aquamarine','pink']
    plt.plot(x,y,'-o',color=colors[i])
    plt.xlabel("Time")
    plt.ylabel("Oddy")
    plt.xlim([5000,8500])
    plt.ylim([-50000,500000])
    plotname = "Oddy Spike for Run:  " + str(i)
    plt.title(plotname)
    plt.show()
```
